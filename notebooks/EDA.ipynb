```python
df.nunique()
```

checking unique values in each column 

```python
categorical_feature=df.select_dtypes(include='object').columns
print(categorical_feature)
print('Unique values in categorical features are : ',end='\n\n')
for feature in categorical_feature:
    print(f'{feature} : {df[feature].unique()}')
```

Finding numerical and categorical features 

```python
num_feature= [feature for feature in df.columns if feature not in categorical_feature]
print(f'Numerical features : {num_feature}')
print(f'categorical_feature : {categorical_feature}')

discrete_feature= [feature for feature in num_feature if df[feature].nunique() < 25]
continuous_feature= [feature for feature in num_feature if feature not in discrete_feature]
print(f'Discrete features : {discrete_feature}')
print(f'Continuous features : {continuous_feature}')
```

Creating new features for total score and average 

```python
df['Total_score']=df['math score']+df['reading score']+df['writing score']
df['average_score']=df['Total_score']/3
df.head()
```
